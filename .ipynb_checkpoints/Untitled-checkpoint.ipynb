```python
import cv2
import mediapipe as mp
import numpy as np
from matplotlib import pyplot as plt
import glob
```

```python
# Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()
```

```python
cap = cv2.VideoCapture('/home/becode2/UCL_Face/data/2022-04-07-105629.webm')
ret, image = cap.read()
height, width, channels = image.shape
name= "out.mp4"
out = cv2.VideoWriter(name, cv2.VideoWriter_fourcc(*"mp4v"), 30, (image.shape[0], image.shape[1]), isColor=False)

landmark_points_68 = [162,234,93,58,172,136,149,148,152,377,378,365,397,288,323,454,389,71,63,105,66,107,336,
                  296,334,293,301,168,197,5,4,75,97,2,326,305,33,160,158,133,153,144,362,385,387,263,373,
                  380,61,39,37,0,267,269,291,405,314,17,84,181,78,82,13,312,308,317,14,87]

while cap.isOpened():
    ret, image = cap.read()
    if ret is not True:
        break
    height, width, channels = image.shape
    # Facial landmarks
    result = face_mesh.process(image)
    print(result)
    canvas = np.zeros((image.shape[1], image.shape[0], 1), dtype = "uint8")
    print('multi-face-landmarks', result.multi_face_landmarks)
    for facial_landmarks in result.multi_face_landmarks:
        for i in range(0, 468):
            pt1 = facial_landmarks.landmark[i]
            x = int(pt1.x * width)
            y = int(pt1.y * height)
            cv2.circle(canvas, (x, y), 3, (100, 100, 0), -1)
    out.write(canvas)

cap.release()
out.release()
```
